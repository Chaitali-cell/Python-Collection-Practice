```python
#Methods
l3=[10,20,30,40,60,70,80,90]
l3
```

```python
l3.append(50) #append operation used to perform add value to end
l3
```

```python
l3.append(50,100) #append operation does not accept more than one value
l3
```

```python
l3.insert(3,500) #insert operation help to add value in between any location
l3
```

```python
l3.insert(4,200,300) #insert operation does not accept more than one value
l3
```

```python
l3.insert(5,[500,700,900])#insert operation taken [] as a single argument
l3
```

```python
l3.extend(500) #extend operation does not perform on integer
l3
```

```python
l3.extend('pune') #string is iterable and taken as a single argument
print(l3)
```

```python
l3.extend([34,67,94,33]) #extend operation taken [] as a single argument and its iterable
print(l3)
```
