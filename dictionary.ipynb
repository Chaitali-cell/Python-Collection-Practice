```python
#print dictionary
d3 = {1:'ram',2:'sham',3:'joy',4:'vijay' } #dictionary is in key value pair
d3
```

```python
d3.keys() #print only keys
```

```python
d3.values() #print only values
```

```python
d3.items() #print key and value pair
```

```python
d3[4]='om' #replace the value using index position
d3
```

```python
d3[5]='ajay' #insert key-value pair 
d3
```

```python
d3.update({6:'sachin'}) #add the key-value
d3
```

```python
d3.popitem() #remove key-value pair using pop function
d3
```

```python
d3.pop(4) #remove key-value pair from the dict at index position 4
d3
```

```python
del d3[2] #delete key-value pair
d3
```
